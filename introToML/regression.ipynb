```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
fastf1.Cache.enable_cache('../f1_cache')
```

```python
session = fastf1.get_session(2021,'Silverstone','R')
session.load()
```

```python
laps = session.laps
race_laps = laps.pick_quicklaps().copy()


race_laps['LapTime_sec'] = race_laps['LapTime'].dt.total_seconds()


ml_data = race_laps[['Driver', 'LapNumber', 'Compound', 'TyreLife', 'LapTime_sec']].dropna()

display(ml_data.head())
```

One-hot Encoding

```python
encoded_laps = pd.get_dummies(ml_data,columns=['Compound','Driver'],prefix=['Compound','Driver'])
display(encoded_laps)
```
